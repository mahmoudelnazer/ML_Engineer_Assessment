```python
import os
import torch
import torch.nn as nn
import numpy as np
from torchvision.models import resnet18, ResNet18_Weights
from PIL import Image
from torch.utils.data import DataLoader
import torch
import torch.nn.functional as F
from torchvision import transforms
from tqdm import tqdm
import os
import random
import shutil
```

```python
from torch.nn.modules.activation import Softmax
class Model(nn.Module):

    def __init__(self):
        super(Model, self).__init__()
        pretrained_model = resnet18(weights=ResNet18_Weights.DEFAULT)
        self.backbone = nn.Sequential(*list(pretrained_model.children())[:-2])
        
        # Initialize the required Layers
        self.have_object = nn.Sequential(
            nn.Conv2d(512, 1024, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            nn.Conv2d(1024, 512, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            nn.Conv2d(512, 256, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            nn.Conv2d(256, 128, kernel_size=3, padding=1),
            nn.AdaptiveAvgPool2d(output_size=1),
            nn.Flatten(),
            nn.Linear(128, 64),
            nn.ReLU(inplace=True),
            nn.Linear(64, 1),
            
        )
        
        self.cat_or_dog = nn.Sequential(
            nn.Conv2d(512, 2048, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            nn.Conv2d(2048, 1024, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            nn.Conv2d(1024, 512, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            nn.Conv2d(512, 256, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            nn.Conv2d(256, 128, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            nn.Conv2d(128, 64, kernel_size=3, padding=1),
            nn.AdaptiveAvgPool2d(output_size=1),
            nn.Flatten(),
            nn.Linear(64, 32),
            nn.ReLU(inplace=True),
            nn.Linear(32, 3),
        )
        
        self.specie = nn.Sequential(
            nn.Conv2d(512, 4096, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            nn.Conv2d(4096, 2048, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            nn.Conv2d(2048, 1024, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            nn.Conv2d(1024, 512, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            nn.Conv2d(512, 256, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            nn.Conv2d(256, 128, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            nn.Conv2d(128, 128, kernel_size=3, padding=1),
            nn.AdaptiveAvgPool2d(output_size=1),
            nn.Flatten(),
            nn.Linear(128, 64),
            nn.ReLU(inplace=True),
            nn.Linear(64, 32),
            nn.ReLU(inplace=True),
            nn.Linear(32, 10),
        )
        
        self.bbox = nn.Sequential(
            nn.Conv2d(512, 1024, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            nn.Conv2d(1024, 512, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            nn.Conv2d(512, 256, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            nn.Conv2d(256, 128, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            nn.Conv2d(128, 4, kernel_size=3, padding=1),
            nn.AdaptiveAvgPool2d(output_size=1),
            nn.Flatten()
        )
        

    def forward(self, input):
        out_backbone = self.backbone(input)
        
        have_object = self.have_object(out_backbone)
        cat_or_dog = self.cat_or_dog(out_backbone)

        specie = self.specie(out_backbone)
        bbox = self.bbox(out_backbone)
        
        return {
            "bbox": bbox,
            "object": have_object,
            "cat_or_dog": cat_or_dog,
            "specie": specie
        }
```

```python

from bs4 import BeautifulSoup

def read_xml_file(path):
    with open(path, 'r') as f:
        data = f.read()
    bs_data = BeautifulSoup(data, 'xml')
    return {
        "cat_or_dog": bs_data.find("name").text,
        "xmin": int(bs_data.find("xmin").text),
        "ymin": int(bs_data.find("ymin").text),
        "xmax": int(bs_data.find("xmax").text),
        "ymax": int(bs_data.find("ymax").text),
        "specie": "_".join(path.split(os.sep)[-1].split("_")[:-1])
    }
```

```python

from torch.utils.data import Dataset
import torchvision.transforms as transforms

class CustomDataset(Dataset):
    def __init__(self, dataset_path, images_list ):
        self.dataset_path = dataset_path
        self.images_list = images_list
        self.transform = transforms.Compose([transforms.Resize((224, 224)), transforms.ToTensor()])
        self.image = None

    def __len__(self):
        return len(self.images_list)

    def __getitem__(self, index):
        name = self.images_list[index].split(".")[0]
        image_path = os.path.join(self.dataset_path, "images", self.images_list[index])
        xml_path = os.path.join(self.dataset_path, "labels", name+".xml")
        if os.path.exists(image_path+'.jpg'):
          self.image = Image.open(image_path+'.jpg').convert('RGB')

        if os.path.exists(image_path+'.jpeg'):
          self.image = Image.open(image_path+'.jpeg').convert('RGB')
        

        if os.path.exists(image_path+'.png'):
          self.image = Image.open(image_path+'.png').convert('RGB')

       
        if self.transform is not None:
            image= self.image
            image = self.transform(image)

        if os.path.exists(xml_path):
            xml_data = read_xml_file(xml_path)
            cat_or_dog = 0 if xml_data['cat_or_dog'] == 'cat' else 1
            xmin = xml_data['xmin']
            ymin = xml_data['ymin']
            xmax = xml_data['xmax']
            ymax = xml_data['ymax']
            specie = labelid_specie[xml_data['specie']]
            object_=1
        else:
            cat_or_dog = 2
            xmin = 0
            ymin = 0
            xmax = 0
            ymax = 0
            specie = 9
            object_=0
        bbox = [xmin, ymin, xmax, ymax]
        return image, (object_, cat_or_dog, bbox, specie)
```

```python
# import our library
import torchmetrics

def train(epochs, train_list, val_list, batch_size, device, model_weights=None):

    # Initialize Model and Optimizer
    model = Model()

    if model_weights is not None:
        
        model.load_state_dict(torch.load(model_weights))

    if torch.cuda.is_available():
        model = model.cuda()

    
    optimizer = torch.optim.Adam([{'params': model.have_object.parameters(), 'lr': 0.001},
                                  {'params': model.specie.parameters(), 'lr': 0.001},
                                  {'params': model.cat_or_dog.parameters(), 'lr': 0.001},
                                  {'params': model.bbox.parameters(), 'lr': 0.001}])

    #Initialize Loss Functions
    have_object_loss = torch.nn.BCEWithLogitsLoss()
    specie_loss      = torch.nn.CrossEntropyLoss()
    cat_or_dog_loss  = torch.nn.CrossEntropyLoss()
    bbox_loss        = torch.nn.MSELoss()  # Not necessary you need to apply function to all coordinates together, You can have separate loss functions for all coordinates too

    training_dataset = CustomDataset("assessment_dataset/", images_list=train_list)
    training_loader = DataLoader(training_dataset, batch_size=batch_size, shuffle=True, num_workers=2)

    best_scores = {'object': 0, 'cat_or_dog': 0, 'specie': 0, 'bbox': 0}
    best_weights = None

    def train_one_epoch(epoch_index, tb_writer):
        running_loss = 0.
        last_loss = 0.
        
        # Here, we use enumerate(training_loader) instead of
        # iter(training_loader) so that we can track the batch
        # index and do some intra-epoch reporting
        for i, data in enumerate(training_loader):
            # Every data instance is an input + label pair
            inputs, labels = data
            inputs = inputs.to(device)

            labels= {'object':labels[0].type(torch.FloatTensor).to(device), 
                     'cat_or_dog': F.one_hot(labels[1], 3).type(torch.FloatTensor).to(device), 
                     'bbox': torch.stack(labels[2]).transpose(0, 1).type(torch.FloatTensor).to(device), 
                     'specie':F.one_hot(labels[3], 10).type(torch.FloatTensor).to(device)}
            
            optimizer.zero_grad()

            # Make predictions for this batch
            outputs = model(inputs)
            object_loss = have_object_loss(outputs["object"].squeeze(), labels["object"])
            specie_loss_ = specie_loss(outputs["specie"], labels["specie"])
            cat_or_dog_loss_ = cat_or_dog_loss(outputs["cat_or_dog"], labels["cat_or_dog"])
            bbox_loss_ = bbox_loss(outputs["bbox"], labels["bbox"])
     
            loss = object_loss + specie_loss_ + cat_or_dog_loss_ + bbox_loss_

            # Backpropagation and optimization
            loss.backward()
            optimizer.step()

            # Gather data and report
            running_loss += loss.item()
            if i % 49 == 0:
                last_loss = running_loss / 49  # loss per batch
                running_loss = 0.
        return last_loss

    for i in range(epochs):
        epoch_loss = train_one_epoch(i, None)
        print(f'Epoch {i} Loss: {epoch_loss}')

        torch.save(model.state_dict(), "model.pth")
        score_object, score_cat_or_dog, score_specie, MAE_bbox = test(model, val_list)
        
        print(f"OD Score : {score_object} & Cat_Dog Score: {score_cat_or_dog} & Species Score: {score_specie}& BBox MAE  : {MAE_bbox}")
        
        print("##################################################################")
        
        # comparing between curent and prev score
        if score_object > best_scores['object'] and score_cat_or_dog > best_scores['cat_or_dog'] \
                and score_specie > best_scores['specie'] and MAE_bbox < best_scores['bbox']:
            best_scores['object'] = score_object
            best_scores['cat_or_dog'] = score_cat_or_dog
            best_scores['specie'] = score_specie
            best_scores['bbox'] = MAE_bbox
            best_weights = model.state_dict().copy()  

    # saving the weights
    torch.save(best_weights, "best_model.pth")

```

```python
def test(model, val_list):
    def post_process_object(x):
      return (x >= 0.5).float()

    def post_process_cat_or_dog(x):
        max_prob_indices = torch.argmax(x, dim=1)
        output = torch.zeros_like(x)
        output.scatter_(1, max_prob_indices.view(-1, 1), 1)
        return output

    def post_process_specie(x):
        max_prob_indices = torch.argmax(x, dim=1)
        output = torch.zeros_like(x)
        output.scatter_(1, max_prob_indices.view(-1, 1), 1)
        return output

    def post_process_bbox(x):
        return x

    val_dataset = CustomDataset("assessment_dataset/", images_list=val_list)
    val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False, num_workers=2)

    metric_object = torchmetrics.classification.BinaryAccuracy().to(device)  
    metric_cat_or_dog = torchmetrics.Accuracy(task='multiclass', num_classes=3).to(device)  
    metric_specie = torchmetrics.Accuracy(task='multiclass', num_classes=10).to(device)
    metric_bbox = torchmetrics.MeanAbsoluteError().to(device)  

    for i, data in enumerate(val_loader):
        inputs, labels = data
        inputs = inputs.to(device)

        labels= {'object':labels[0].type(torch.FloatTensor).to(device), 
                 'cat_or_dog': F.one_hot(labels[1], 3).type(torch.FloatTensor).to(device), 
                 'bbox': torch.stack(labels[2]).transpose(0, 1).type(torch.FloatTensor).to(device), 
                 'specie':F.one_hot(labels[3], 10).type(torch.FloatTensor).to(device)}

        # Make predictions for this batch
        outputs = model(inputs)
        
        # Post-process the outputs
        pred_object = post_process_object(outputs["object"].squeeze())
        pred_cat_or_dog = post_process_cat_or_dog(outputs["cat_or_dog"])
        pred_specie = post_process_specie(outputs["specie"])
        pred_bbox = post_process_bbox(outputs["bbox"])

        # Update the metrics
        metric_object.update(pred_object, labels["object"].squeeze())
        metric_cat_or_dog.update(pred_cat_or_dog, labels["cat_or_dog"])
        metric_specie.update(pred_specie, labels["specie"])
        metric_bbox.update(pred_bbox, labels["bbox"])


    score_object = metric_object.compute().item()
    score_cat_or_dog = metric_cat_or_dog.compute().item()
    score_specie = metric_specie.compute().item()
    score_bbox = metric_bbox.compute().item()

    return score_object, score_cat_or_dog, score_specie, score_bbox
```

```python
def draw_bounding_box(image, bbox,imgName, savePath):
  
    pil_image= image
    
    # Create an ImageDraw object
    draw = ImageDraw.Draw(pil_image)

    # Draw the bounding box rectangle
    draw.rectangle(bbox, outline='red', width=2)

    # Convert the PIL Image back to a numpy array
    image_with_bbox = np.array(pil_image)
    r= Image.fromarray(image_with_bbox)
    r.save(os.path.join(savePath, imgName))
    return 
```

```python
def prediction(model_weights,image_folder_path):
  model = Model()
  model.load_state_dict(torch.load(model_weights))
  model.eval()
  
  transform = transforms.Compose([transforms.Resize((224, 224)),transforms.ToTensor()])
  imgList, pilList=[], []

  image= None
  for imgName in os.listdir(image_folder_path):
      img_path= os.path.join(image_folder_path, imgName)
      image = Image.open(img_path).convert('RGB')
      pilList.append(image)
      image_ = transform(image)
      imgList.append(image_)
  imgsBatch= torch.stack(imgList,dim=0)
  
  return model(imgsBatch),pilList
```

```python
from PIL import ImageDraw, Image

def visualize(model_weights, image_folder_path, output_folder="output"):
  def post_process_object(x):
    return (x >= 0.5).float()

  def post_process_cat_or_dog(x):
      max_prob_indices = torch.argmax(x, dim=1)
      output = torch.zeros_like(x)
      output.scatter_(1, max_prob_indices.view(-1, 1), 1)
      return output

  def post_process_specie(x):
      max_prob_indices = torch.argmax(x, dim=1)
      output = torch.zeros_like(x)
      output.scatter_(1, max_prob_indices.view(-1, 1), 1)
      return output

  def post_process_bbox(x):
      return x
  if not os.path.exists(output_folder):
    os.makedirs(output_folder)
  outputs,pilList = prediction(model_weights,image_folder_path)
 
  # Post-process the outputs
  #Object Output
  pred_object = post_process_object(outputs["object"].squeeze())
  pred_objectList = ["True" if i == 1 else "False" for i in list(pred_object)]
  #Specie Output
  pred_specie = post_process_specie(outputs["specie"])
  pred_specieList = [idlabel_specie.get(i.tolist().index(1)) for i in list(pred_specie)]
  #cat_or_dog Output
  pred_cat_or_dog = post_process_cat_or_dog(outputs["cat_or_dog"])
  pred_cat_or_dogList = [idlabel_catDog.get(i.tolist().index(1)) for i in list(pred_cat_or_dog)]
  #bbox Output
  pred_bbox = post_process_bbox(outputs["bbox"])

  tupList= list(zip(pred_objectList,pred_cat_or_dogList, pred_specieList, pred_bbox.detach().numpy().tolist()))
  dictList={}
  for i, tup in enumerate(tupList):
    obj, cat_dog, specie, bbox= tup
    if obj =='True':
      dictList[os.listdir(image_folder_path)[i]]= { "has_object": True, "cat_or_dog": cat_dog, "specie": specie, "xmin": bbox[0], "ymin": bbox[1], "xmax": bbox[2], "ymax": bbox[3] }
      draw_bounding_box(pilList[i], bbox,os.listdir(image_folder_path)[i], output_folder) 

    else:
      dictList[os.listdir(image_folder_path)[i]]= { "has_object": False, "cat_or_dog": 'NA', "specie": "NA", "xmin": 0, "ymin": 0, "xmax": 0, "ymax": 0 } 
      pilList[i].save(os.path.join(output_folder, os.listdir(image_folder_path)[i]))
  return dictList 
```
